## Opening FITS files and loading the image data

Let's open the g-band FITS file and find out what it contains.

### SHORTCUT

If you don't need to examine the FITS header, you can call `fits.getdata` to bypass the previous steps.

```python
image_u = fits.getdata("frame-u-006073-4-0063.fits")
print(type(image_u))
print(image_u.shape)

image_g = fits.getdata("frame-g-006073-4-0063.fits")
print(type(image_g))
print(image_g.shape)

image_r = fits.getdata("frame-r-006073-4-0063.fits")
print(type(image_r))
print(image_r.shape)
```

## Viewing the image data and getting basic statistics

Let's get some basic statistics about our image

```python
for image in [image_u, image_g, image_r]:
    print('Min:', np.min(image))
    print('Max:', np.max(image))
    print('Mean:', np.mean(image))
    print('Stdev:', np.std(image))
    print 
```

```python
plt.imshow(image_g/image_r, origin='lower', cmap='afmhot', norm=LogNorm())
```

```python
combo = np.dstack((image_r, image_g, image_u))
combo.shape
```

```python
plt.imshow(np.log10(np.abs(combo)), origin='lower')
```
