```python
%matplotlib inline 
import seaborn as sns
import numpy as np
import astropy
from astropy.cosmology import Planck13
from matplotlib import pyplot as plt
sns.set_context('poster')
```

```python
SNdata = np.loadtxt('data/SNdata.txt')
SNdata = SNdata[SNdata[:,0].argsort()] #This sorts SNdata by the values in column 0.
z_data = SNdata[:,0]
DM_data = SNdata[:,1]
DM_err = SNdata[:,2]
```

# We'll reuse our cosmological model from the homework 3 solution, with some small changes.

### In order to find the $\chi^2$ goodness-of-fit, we'll have to modify the equations below to make H$_0$ a variable, instead of a globally defined parameter.

### Again we define our cosmological model from David Hogg's [arXiv paper](http://www.as.utexas.edu/~gebhardt/a376f14/hogg.pdf) with $\Omega_M$, $\Omega_k$, and $\Omega_\Lambda$ as variables.  *But now, we include H$_0$ as a variable.*

Since all dependence on H$_0$ in our model comes in through the Hubble distance, we'll go ahead and make a function for it: $$D_H = \frac{c}{H_0}.$$  

```python
#Hogg eq. 4: Hubble Distance
def DH(H0):
    # Assume H0 is in km/s/Mpc
    c = 2.99792e5 # Use speed of light in km/s for simplicity
    dh = c / H0
    dh = 1e6 * dh #convert from Mpc to pc
    return dh
```

```python
# Hogg eq. 14: E(z)
def E(z, omega_matter, omega_k, omega_lambda):
    return np.sqrt(omega_matter*(1+z)**3 + omega_k*(1+z)**2 + omega_lambda)

from scipy import integrate
# Hogg eq. 15: Line-of-sight Comoving Distance
def Dcom_L(z, omega_matter, omega_k, omega_lambda, H0):
    # Define a function of only one variable for passing into integrate.quad()
    # There are other, more concise ways to do this, but I think this is 
    # the easiest to follow.
    def integrand(x):
        return DH(H0) / E(x, omega_matter, omega_k, omega_lambda)
    # Now we perform the integral for every entry in z, and add it to Dlist
    Dlist = []
    for redshift in z:
        integral, err = integrate.quad(integrand, 0, redshift)
        Dlist.append(integral)
    return np.asarray(Dlist) 
    
# Hogg eq. 16: Transverse Comoving Distance
def Dcom_T(z, omega_matter, omega_k, omega_lambda, H0):
    dc = Dcom_L(z, omega_matter, omega_k, omega_lambda, H0)
    if omega_k == 0:
        return dc
    else:
        if omega_k > 0:
            sqOk = np.sqrt(omega_k)
            return DH(H0) * np.sinh(sqOk * dc / DH(H0)) / sqOk
        else:
            sqOk = np.sqrt(-omega_k)
            return DH(H0) * np.sin(sqOk * dc / DH(H0)) / sqOk
        
# Hogg eq. 20: Luminosity Distance
def DL(z, omega_matter, omega_k, omega_lambda, H0):
    return Dcom_T(z, omega_matter, omega_k, omega_lambda, H0) * (1+z)

# Hogg eq. 25: DISTANCE MODULUS
def DM(z, omega_matter, omega_k, omega_lambda, H0):
    return 5 * np.log10(DL(z, omega_matter, omega_k, omega_lambda, H0) / 10) 
```

# Now let's find the $\chi^2$ goodness-of-fit for our full model:

To do this, we need to calculate the model-dependent $\chi^2$ distribution for our data against each model:
$$\chi^2 = \sum_{i=1}^n \left( \frac{X_i - \mu_i}{\sigma_i} \right)^2.$$
Here, $X_i$ is an observed data point, $\mu_i$ what the model would predict at that point, and $\sigma_i$ is the error on the observed data point. **To clarify, $X_i - \mu_i$ is the residual you calculated in homework 3.**


For example, the $\chi^2$ fit for this dataset against the Planck13 model would be the following:

```python
standard = DM(z_data, Planck13.Om0, Planck13.Ok0, Planck13.Ode0, Planck13.H0.value)
residual = DM_data - standard
chi2 = np.sum((residual/DM_err)**2)
chi2
```

So the $\chi^2$ goodness-of-fit for the Planck13 standard model against our dataset is 1055.  That's a pretty lousy fit, considering that a good fit results in a $\chi^2$ value about equal to the number of data points (288 here).

Now we need to find the **_best_** fit. This is fairly easy to do for a single parameter: Hold all the other variable fixed, change only the one you're interested in, and see which one gives the smallest result.

```python
Omchi2 = []
OmVals = np.linspace(0,1)
for om in OmVals:
    model = DM(z_data, om, Planck13.Ok0, Planck13.Ode0, Planck13.H0.value)
    fit = np.sum((DM_data-model)**2 / DM_err**2)
    Omchi2.append(fit)
Omchi2 = np.asarray(Omchi2)
#find the index of the smallest value for chi^2
ix_min = np.argmin(Omchi2)
# That index corresponds to the best fit for Omega_M
print OmVals[ix_min]
```

We can plot this to verify:

```python
plt.plot(np.linspace(0,1), Omchi2)
plt.axvline(OmVals[ix_min], linestyle='--', color='grey')
plt.axhline(Omchi2[ix_min], linestyle='--', color='grey')
plt.xlabel('$\Omega_M$')
plt.ylabel('$\chi^2$')
```

# But, there's a problem with this approach:

We want to find the value that provides the best fit for **_ALL_** the parameters _**at the same time.**_  Think of it like being an ant crawling along the inside of a bowl.  If you walk in a straight line, you'll find one point that is lower than all the other points on your journey.  BUT, unless your trip takes you through the exact center of the bowl, you won't actually find the global minimum.  

Insteading of finding the value of 'x' for which 'y' is at a minimum, we need to find the combination of values 'x, y' for which 'z' is at a minimum. So, we need to consider two variable at once.  

Luckily, we're going to assume $\Omega_k = 0$, meaning that $\Omega_\Lambda = 1 - \Omega_M$, so we really only have to worry about two independent variables: $\Omega_M$ and H$_0$.

What we need is a 2d grid of points, with $\Omega_M$ along one axis, and H$_0$ along the other.  This bit of code does precisely that:

```python
# Create a 2d grid of datapoints for Omega_M and H_0
# For Omega_M, go from 0 to 1 with point spacing of .1 --> 0:1:.1
# For H_0, go from 60 to 80 with point spacing of .2 --> 60:80:2
# This results in a 10 by 10 grid.
grid_Om, grid_H0 = np.mgrid[0:1:.1, 60:80:2]

#After doing this once and getting a feel for where the center is, narrow the range and increase the resolution:
grid_Om, grid_H0 = np.mgrid[.3:.5:.002, 62:65:.03]
grid_Om.shape
```

Now we need to create a 2d grid to store the $\chi^2$ value for each point:

```python
grid_chi2 = np.zeros_like(grid_Om)
```

Then, for each x,y point, calculate $\chi^2$.  With a 100 by 100 grid, this loops 10000 times, so expect it to take a while.

```python
for i in xrange(grid_Om.shape[0]):
    for j in xrange(grid_Om.shape[1]):
        model = DM(z_data, grid_Om[i,j], 0, 1.-grid_Om[i,j], grid_H0[i,j])
        grid_chi2[i,j] = np.sum((DM_data-model)**2 / DM_err**2)
```

It's now trivial to find our minimum $\chi^2$ fit:

```python
chi2min = np.min(grid_chi2)
chi2min
```

```python
global_min = np.where(grid_chi2 == chi2min)
print global_min
```

And the best fit values for $\Omega_M$ and H$_0$ are:

```python
print grid_Om[global_min], grid_H0[global_min]
```

### Recall that the 68% confidence limit is where $\chi^2$ has increased by 1, so a contour plot of $\chi^2 - \chi^2_{\rm min}$ makes it easy to see precisely where the confidence limit is.

```python
CS = plt.contour(grid_Om, grid_H0, grid_chi2-chi2min, levels=[1, 10], colors='k')
plt.clabel(CS, inline=1, fontsize=10)
plt.axvline(grid_Om[global_min], linestyle='--', color='grey')
plt.axhline(grid_H0[global_min], linestyle='--', color='grey')
plt.xlabel('$\Omega_M$')
plt.ylabel(r'$H_0$ [${\rm km / s / Mpc}$]')
```

###### You can see from this that the 68% confidence limit is an off-axis ellipse, so we lose some information when we boil this down to simple confidence intervals for $\Omega_M$ and H$_0$.  To find these values though, we can do the following:

The 68% confidence limits are intertwined; each requires some knowledge of the other.  To find the 68% limit for $\Omega_M$, for example, we take the $\chi^2$ values for various $\Omega_M$ models assuming H$_0$ is fixed at the global minimum:

```python
Om_min_ix = global_min[0][0]
H0_min_ix = global_min[1][0]
```

```python
OmLimits = np.where(grid_chi2[:, H0_min_ix] - chi2min < 1.)[0]
H0Limits = np.where(grid_chi2[Om_min_ix, :] - chi2min < 1.)[0]
```

```python
print grid_Om[OmLimits, H0_min_ix]
print grid_H0[Om_min_ix, H0Limits]
```

###### So, the 68% confidence interval for $\Omega_M$ is 0.392 to 0.422, and the limits for H$_0$ are 63.44 to 63.8

And just for fun, let's see how these parameters compare to the standard model and our dataset.

```python
plt.errorbar(z_data, DM_data, yerr=DM_err, color='grey', ls='none', label='SN 1A data')
plt.plot(z_data, DM(z_data, Planck13.Om0, Planck13.Ok0, Planck13.Ode0, Planck13.H0.value), label='Planck 2013 best fit')
plt.plot(z_data, DM(z_data, 0.406, 0, 1-.406, 63.62), label='$\Omega_M = 0.406$, H$_0 = 63.62$')

plt.xlabel('Redshift')
plt.ylabel('Distance Modulus')
plt.legend(loc=4)
```

And now the residuals:

```python
standard = DM(z_data, 0.406, 0, 1-.406, 63.62)
plt.scatter(z_data, DM_data-standard, label='SN 1A Residuals', color='grey', s=7)
plt.plot(z_data, DM(z_data, Planck13.Om0, Planck13.Ok0, Planck13.Ode0, Planck13.H0.value)-standard,
        label='Planck 2013 best fit')
plt.plot(z_data, np.zeros_like(z_data), label='$\Omega_M = 0.406$, H$_0 = 63.62$')

plt.xlabel('Redshift')
plt.ylabel('Residuals')
plt.xlim(0,1.6)
plt.legend(loc=4)
```
